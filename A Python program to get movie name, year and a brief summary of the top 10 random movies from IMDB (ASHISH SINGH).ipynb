```python
#importing all the libraries 
import requests
from bs4 import BeautifulSoup
import pandas as pd

#Sending request to the server to download the site’s content.
url='https://m.imdb.com/chart/top'
r=requests.get(url)

#print(r.content) [just for cheking]

soup=BeautifulSoup(r.content,'html5lib')

#ccreating empity lists that we will fill this later
movie_name = []
year = []
time = []
rating = []

movie_data = soup.findAll('div', attrs= {'class': 'row chart-row'})

for store in movie_data:
    name = store.img['alt']
    movie_name.append(name)
   # print(movie_name)
  
    year_of_release = store.h4.findAll('span', class_ = 'unbold')[1].text.replace('(', '').replace(')', '')
    year.append(year_of_release)
    #print(year)

    rate = store.find('span', class_ = 'imdb-rating').text.replace('\n', '')
    rating.append(rate)
    #rint(rating)

    #saving all the informations 
movie_DF = pd.DataFrame({'Name of movie':     movie_name,'        Year of relase': year,      'rating':rating})
 
#LETS PRINT TOP 10 RANDOM MOVIES WITH THEIR YEAR OF RELASE AND RATINGS

    
print(movie_DF.sample(10))
# TOP 250 MOVIES
print(movie_DF)

```
